## Dataset

**Titanic Dataset:** The dataset for the Titanic Survival Prediction project, sourced from Kaggle, is a well-known dataset often used in data science and machine learning competitions. It provides detailed information about the passengers aboard the RMS Titanic, offering insights into their demographics, travel class, and survival status. The dataset is composed of 12 features, each contributing valuable data that can be analyzed to predict survival outcomes. Below is an explanation of the key features:


**PassengerId:** A unique identifier assigned to each passenger. This feature is primarily used to distinguish between passengers and has no direct impact on the survival prediction.

**Survived:** This is the target variable for the project, indicating whether a passenger survived (1) or did not survive (0). The goal is to predict this outcome using the other features in the dataset.

**Pclass:** This feature represents the ticket class of the passenger, with values 1, 2, and 3 corresponding to first, second, and third class, respectively. Pclass is a proxy for socio-economic status, with first-class passengers generally having better survival rates.

**Name:** The full name of the passenger, which includes titles (e.g., Mr., Mrs., Miss). While this feature contains textual data, it can be used for feature engineering, such as extracting titles that might correlate with age, gender, and survival.

**Sex:** The gender of the passenger (male or female). Gender is a crucial predictor, as survival rates varied significantly between men and women due to the "women and children first" policy during the evacuation.

**Age:** The age of the passenger in years. This feature is important as age played a role in survival, with younger children and women generally having higher survival rates. Missing values in this feature are often handled through imputation.

**SibSp:** The number of siblings and/or spouses aboard the Titanic with the passenger. This feature provides insight into family size and relationships, which could influence survival due to family members assisting one another.

**Parch:** The number of parents and/or children aboard the Titanic with the passenger. Like SibSp, this feature helps understand the family structure, which might impact the likelihood of survival.

**Ticket:** The ticket number assigned to the passenger. While this feature is categorical and non-numerical, it could hold some hidden patterns, such as grouping passengers who traveled together.

**Fare:** The amount paid for the ticket, which can be used as an indicator of wealth and social status. Higher fares are generally associated with first-class passengers, who had higher survival rates.

**Cabin:** The cabin number assigned to the passenger. This feature has many missing values, but when available, it can provide information about the passenger's location on the ship, which might be related to survival.

**Embarked:** The port of embarkation where the passenger boarded the Titanic. The values are C for Cherbourg, Q for Queenstown, and S for Southampton. This feature can offer insights into the passengers' origin and, potentially, their socio-economic status.

## Feature Selection
Feature selection is the process of identifying the most relevant features in a dataset that contribute to the accuracy and performance of a machine learning model. For the Titanic Survival Prediction project, feature selection is crucial because it helps in building a simpler, faster, and more interpretable model by eliminating irrelevant or redundant features.

#### Steps in Feature Selection:

**Initial Analysis and Understanding:** The first step involves understanding the dataset and its features. Some features may be immediately recognized as irrelevant or non-contributory to the prediction task. For example, PassengerId is a unique identifier and does not provide any useful information for predicting survival.

**Correlation Analysis:** A correlation matrix is used to identify the relationships between numerical features and the target variable (Survived). Features with a high correlation to the target variable are considered more important. However, features with high inter-correlation (multicollinearity) may be redundant and could be removed.
 
**Feature Importance from Models:** Machine learning models like Random Forest, Gradient Boosting, or Decision Trees can provide insights into feature importance. These models rank features based on their contribution to the prediction, helping to identify which features should be retained.

**Feature Engineering:** Before finalizing the selection, new features can be created to enhance the predictive power of the model. For example, combining SibSp and Parch into a single FamilySize feature or extracting titles from the Name feature may lead to better predictions.

**Dimensionality Reduction Techniques:** Techniques like Principal Component Analysis (PCA) can be used to reduce the dimensionality of the dataset, especially if there are many features. PCA transforms the data into a lower-dimensional space while retaining as much variance as possible.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.impute import SimpleImputer
```

```python
# Separate Features (X) and Target Variable (y)
X = titanic_df.drop('Survived', axis=1)
y = titanic_df['Survived']
```

```python
# Handle Missing Values using Mean Imputation
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
# Initialize RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Fit the RandomForestClassifier to Estimate Feature Importance
rf_clf.fit(X_imputed, y)
```

```python
# Extract Feature Importance Scores
feature_importances = rf_clf.feature_importances_
```

```python
# Select Features with Importance Scores above a Certain Threshold
selector = SelectFromModel(rf_clf, threshold=0.05)
selector.fit(X_imputed, y)
```

```python
# Get the Indices of Selected Features
selected_features_indices = selector.get_support(indices=True)
```

```python
# Get the Names of Selected Features
selected_features_names = X.columns[selected_features_indices]
```

```python
# Display Selected Features
print("Selected Features:")
print(selected_features_names)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X[selected_features_names], y, test_size=0.2, random_state=42)
```

```python
# Print the Shapes of the Training and Testing Sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
# Initialize Instances of the Chosen Algorithms with Default Parameters
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}
```

```python
# Preprocess Features if Necessary (E.g., Standardization for SVM or KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train and Evaluate Each Model using Cross-Validation
for name, model in models.items():
    print(f"Training {name}...")

    # Perform Grid Search for HyperParameter Tuning
    if name in ['Random Forest', 'Gradient Boosting', 'Support Vector Machine', 'K-Nearest Neighbors']:
        param_grid = {}  # Define HyperParameter Grid for Each Model
        if name == 'Random Forest':
            param_grid = {'n_estimators': [100, 200, 300],
                          'max_depth': [None, 5, 10],
                          'min_samples_split': [2, 5, 10]}
        elif name == 'Gradient Boosting':
            param_grid = {'n_estimators': [100, 200, 300],
                          'learning_rate': [0.01, 0.1, 0.5],
                          'max_depth': [3, 5, 7]}
        elif name == 'Support Vector Machine':
            param_grid = {'C': [0.1, 1, 10],
                          'kernel': ['linear', 'rbf', 'poly']}
        elif name == 'K-Nearest Neighbors':
            param_grid = {'n_neighbors': [3, 5, 7],
                          'weights': ['uniform', 'distance']}
        
        grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
        grid_search.fit(X_train_scaled if name != 'K-Nearest Neighbors' else X_train, y_train)

        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_test_scaled if name != 'K-Nearest Neighbors' else X_test)
    else:
        model.fit(X_train_scaled if name != 'K-Nearest Neighbors' else X_train, y_train)
        y_pred = model.predict(X_test_scaled if name != 'K-Nearest Neighbors' else X_test)
```

```python
    # Evaluate Model Performance
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {name}: {accuracy:.2f}")
```

```python
    # Print Classification Report and Confusion Matrix
    print(f"Classification Report of {name}:")
    print(classification_report(y_test, y_pred))
    print(f"Confusion Matrix of {name}:")
    print(confusion_matrix(y_test, y_pred))
    print()
```

## HyperParameter Tuning
Hyperparameter tuning is the process of optimizing the parameters of a machine learning model that are not learned from the data but set before the learning process begins. These parameters, known as hyperparameters, have a significant impact on the modelâ€™s performance. Properly tuning them ensures that the model generalizes well to new data and performs optimally.

These vary by model but often include parameters like learning rate, depth of trees, number of trees (in ensemble models), and regularization strength.

#### Tuning Techniques:

**Grid Search:** Systematically explores a predefined range of hyperparameter values.

**Random Search:** Randomly samples hyperparameter combinations to find a good set.

**Bayesian Optimization:** Uses past results to guide the search more intelligently.


##### Evaluation:
Cross-validation is often used during tuning to ensure the model generalizes well across different data splits.

##### Outcome: 
The result is a model that performs better on unseen data by using the optimal hyperparameters found during the tuning process.
